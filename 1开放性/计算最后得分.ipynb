```python
# 计算月销售量
import pandas as pd
from datetime import datetime
import time
```

```python
kaifangScore=pd.DataFrame(columns=['user_id','1','2','3','4','score'])
kaifangScore['user_id']=range(1,98925)
```

```python
kaifangScore
```

```python
# 反向记分，复购率越低越开放性越高
liulan = pd.read_csv('./data/action_othesAttracted.csv')
maxAttrated = liulan['attracted'].max()
for row in liulan.itertuples():
    kaifangScore.loc[row.user_id-1, '1'] = row.attracted/maxAttrated*100
```

```python
# 反向记分，复购率越低越开放性越高
fugou = pd.read_csv('./data/repurchase.csv')
for row in fugou.itertuples():
    kaifangScore.loc[row.user_id-1, '2'] = 100-row.repurchaserate*100
```

```python
jiage = pd.read_csv('./data/priceAll.csv')
maxAvg = jiage['avg'].max()
for row in jiage.itertuples():
    kaifangScore.loc[row.user_id-1, '3'] = row.avg/maxAvg*100
```

```python
xiaoshouliang = pd.read_csv('./data/salesAvg.csv')
maxSales=xiaoshouliang['avgsales'].max()
for row in xiaoshouliang.itertuples():
    kaifangScore.loc[row.user_id-1, '4'] = row.avgsales/maxSales*100
```

```python
for row in kaifangScore.itertuples():
    kaifangScore.loc[row.user_id-1, 'score']=sum(row[2:6])/4
```
